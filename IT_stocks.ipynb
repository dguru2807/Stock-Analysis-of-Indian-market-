```python
#analysis of returns of four major IT companies in India with data obtained from yahoo finance
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
cf.go_offline()
                   
```

```python
import pandas as pd
from pandas_datareader import data

symbol = 'INFY.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dfinfy = data.DataReader(symbol, data_source, start_date, end_date)

```

```python
symbol = 'TCS.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dftcs = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
symbol = 'TECHM.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dftechm = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
symbol = 'WIPRO.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dfwipro = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
symbol = 'HCLTECH.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dfhcl = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
df = pd.concat([dfinfy,dftcs,dftechm,dfwipro,dfhcl],keys=['INFOSYS','TCS','TECHM','WIPRO','HCL'],names=['CompanyName','Date'])
df.head()
```

```python

##df.loc[(['SBI','HDFC','ICICI','AXIS'],['2009-01-02','2020-04-09']),['Close','Volume']]
df1 = df.reset_index( drop=False)

df1.head()
```

```python
# this is for seaborn plots. 

import matplotlib.pyplot as plt
sns.set_style("darkgrid")
plt.figure(figsize=(16, 6))
bankplot=sns.lineplot(x='Date',y='Close',hue='CompanyName', data=df1 )
bankplot.set_title('IT Stocks')
bankplot
```

```python
#this is for plotly plots

df1.iplot(kind='line',x='Date',y='Close',categories='CompanyName',mode='lines')
```
