```python
#analysis of returns of major auto companies in India with data obtained from yahoo finance
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
cf.go_offline()
                   
```

```python
import pandas as pd
from pandas_datareader import data

symbol = 'M&M.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dfmm = data.DataReader(symbol, data_source, start_date, end_date)

```

```python
symbol = 'HEROMOTOCO.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dfhero = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
symbol = 'EICHERMOT.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dfeicher = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
symbol = 'BAJAJ-AUTO.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dfbajaj = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
symbol = 'MARUTI.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dfmaruti = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
symbol = 'TATAMOTORS.NS'
data_source='yahoo'
start_date = '2009-01-01'
end_date = '2019-12-31'

dftata = data.DataReader(symbol, data_source, start_date, end_date)
```

```python
df = pd.concat([dfmm,dfhero,dfeicher,dfbajaj,dfmaruti,dftata],keys=['M&M','HERO','EICHER','BAJAJ','MARUTI','TATA'],names=['CompanyName','Date'])
df.head()
```

```python

##df.loc[(['SBI','HDFC','ICICI','AXIS'],['2009-01-02','2020-04-09']),['Close','Volume']]
df1 = df.reset_index( drop=False)

df1.head()
```

```python
# this is for seaborn plots. 

import matplotlib.pyplot as plt
sns.set_style("darkgrid")
plt.figure(figsize=(16, 6))
bankplot=sns.lineplot(x='Date',y='Close',hue='CompanyName', data=df1 )
bankplot.set_title('Auto Stocks')
bankplot
```

```python
#this is for plotly plots

df1.iplot(kind='line',x='Date',y='Close',categories='CompanyName',mode='lines')
```
